```python
def box_plot(cols):
    fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(15, 10))
    axes = axes.flatten()
    
    for i, col in enumerate(cols):
        sns.boxplot(x=df[col], ax=axes[i])
        axes[i].set_title(f'Box Plot of {col}')
        
    plt.tight_layout()
    plt.show()
    
box_plot(numerical_cols)
        
```
